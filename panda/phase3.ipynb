```python
import numpy as np 
  
import pandas as pd 

```

```python
employees = pd.DataFrame({
    'employee_id':[1,2,3,4,5],
    'name':['Anna', 'Peter','Linda','Bob','James'],
    'department':['IT','HR','Finance','IT','HR']
})

salaries = pd.DataFrame({
    'employee_id':[1,2,3,6,7],
    'salary':[60000,50000,70000,80000,45000],
    'bonus':[5000,10000,7000,8000,4500]
})
```

```python
employees
```

```python
salaries
```

### Merging data (Joins)

```python
# how = 'inner' by default
pd.merge(employees, salaries, on='employee_id', how='inner')
```

```python
# outer join
pd.merge(employees, salaries, on='employee_id', how='outer')
```

```python
# left join
pd.merge(employees,salaries, on = 'employee_id', how = 'left')
```

```python
# right join
pd.merge(employees,salaries,on = 'employee_id',how = 'right')
```

Concatination of two dataframe

```python
df1 = pd.DataFrame({
    'A':['A0','A1','A2'],
    'B':['B0','B1','B2'],
    'C':['C0','C1','C2']
})

df2 = pd.DataFrame({
    'A':['A3','A4','A5'],
    'B':['B3','B4','B5'],
    'C':['C3','C4','C5']
})
```

```python
pd.concat([df1,df2]) # concatenate row wise
```

```python
pd.concat([df1,df2],axis = 1) # concatenate column wise
```

Joining to dataframe

```python
df1 = pd.DataFrame({
    'name':['Alice','Bob','Charlie'],
},
index= [1,2,3])

df2 = pd.DataFrame({
    'score':[87,92,78]
},index = [2,3,4])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2) # joins based on index
```

```python
df1.join(df2, how = 'outer')
```
