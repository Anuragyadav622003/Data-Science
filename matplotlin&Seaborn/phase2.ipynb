```python
import seaborn as sns

import numpy as np  

import matplotlib.pyplot as plt 
```

We use distribution plots to visualize the distribution of quantitative data.

Let's disuss some plots that allow us to visualize the distribution of a data set. These plots are.

displot/histplot

jointplot

pairplot

rugplot

Load Seaborn dataset


```python
df = sns.load_dataset('tips')
df
```

```python
sns.displot(df['total_bill'],bins = 10)
```

```python
# OR
''' kde is kernel density estimater is used to show the distribution of data by a smooth curve 
and by default it is false'''

sns.histplot(df['total_bill'],bins = 5, kde= True) 
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins = 20, kde= True) 
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins = 20,kde = True)
```

```python
# jointplot (two quantities)

sns.jointplot(x = 'total_bill',y = 'tip', data=df,kind='reg' )
# kind = 'hex' or 'scatter' or 'kde' or 'reg'  or 'resid' 
```

```python
sns.pairplot(df)# all pairwise relationships in the dataset(numerical data)
```

```python
sns.pairplot(df,hue = 'sex',palette='rainbow')
# hue is used to add another dimension to plot ( add categorical data)
```

```python
sns.rugplot(df)
```

Now let's discuss using seaborn to plot categorical data! There are few main plot types for this:

boxplot

countplot

```python
sns.countplot(df['sex']) # by default it take y axis as categorical data
```

```python
sns.countplot(x = df['sex'])
```

```python
sns.barplot(x=df['sex'],y = df['total_bill'])
#y (optional) by default is quantitative data and x is qualitative data
```

```python
sns.boxplot(x='tip', y='day', data=df, color = 'skyblue')
```

### Boxplot and Violinplot

A **boxplot** is a chart that shows how data is spread out. It displays the minimum, first quartile (25%), median (50%), third quartile (75%), and maximum values. Outliers are shown as dots outside the "whiskers." This helps you quickly see where most data points are and if there are any unusual values.

A **violinplot** is similar to a boxplot but also shows the shape of the dataâ€™s distribution. It combines a boxplot with a rotated density plot (like a smoothed histogram), so you can see if the data is skewed or has multiple peaks. This makes it easier to understand how the data is distributed.





```python
sns.violinplot(x='day',y = 'total_bill',data=df)
```

### Strip Plot

A **strip plot** is a scatter plot for categorical data. Each point represents an individual data value, plotted along a single axis based on its category. It's useful for showing all data points, but can look cluttered if many points overlap.

### Swarm Plot

A **swarm plot** improves on the strip plot by spreading out overlapping points. This creates a "beeswarm" effect, making it easier to see the distribution and density of the data. Swarm plots are especially helpful for small datasets, as they show individual values clearly.

```python
sns.stripplot(x='day', y = 'total_bill', data = df, jitter=True)
```

```python
sns.swarmplot( x= 'day', y = 'total_bill', data = df)
```

```python
# combine swarm and voilin plot
sns.violinplot(x = 'tip', y = 'day', data = df)
sns.swarmplot(x = 'tip', y = 'day', data = df)
```
