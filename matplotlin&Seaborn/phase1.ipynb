```python

import numpy as np

import pandas as pd      

import matplotlib.pyplot as plt

```

Matplotlib Basic


```python
x = np.linspace(0,5,11)
x
```

```python
y = x**2
y
```

```python
plt.plot(x,y) # can be 3D also 
plt.title('Hello')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x,np.exp(x))
plt.subplot(2,2,3)
plt.plot(x,np.sin(x))
plt.subplot(2,2,4)
plt.plot(x,np.cos(x))
```

```python
fig = plt.figure()
fig.add_axes([0.1,0.1,0.8,0.8]) #([left,bottom,width,height]) in percentage
```

```python
fig = plt.figure()
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis1 = fig.add_axes([0.4,0.3,0.5,0.5])
axis1.plot(x,y)
```

Matplotlib allows the aspect ratio, DPI and figure size to specified when the figure object is created.You can use the figsize and dpi keyword arguments.

figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch (pixel per inch).

```python
fig = plt.figure(figsize=(5,5),dpi = 100)
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis1 = fig.add_axes([0.4,0.3,0.5,0.5])
axis1.plot(x,y)
```

```python
data = np.random.randn(1000)   # 1000 random values from normal distribution

# histogram
plt.hist(data, bins=20, edgecolor='black')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram Example")
plt.show()
```

```python
plt.boxplot(data)
plt.title("Boxplot Example")
plt.ylabel("Values")
plt.show()
```

Saving Plots

```python

x = np.linspace(0, 10, 100)
y = np.sin(x)

fig = plt.figure()

# add an axes object [left, bottom, width, height]
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

# plot on that axes
ax.plot(x, y)

# add labels and title
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_title("Example of ax.plot()")

# save before show
plt.savefig("fig1.png", dpi=900)   # specify extension, e.g., .png/.jpg/.pdf
plt.show() # render and clear the figure so first save and then show
```

Optional Design

```python
plt.plot(x,y,color='purple')
```

Set a Style

```python
plt.plot(x, y, color="red", linewidth=2, linestyle="--", marker="o", label="sin(x)")
plt.xlabel("X-axis", fontsize=12, color="blue")
plt.ylabel("Y-axis", fontsize=12, color="blue")
plt.title("Styled Plot", fontsize=16, fontweight="bold")
plt.legend()
plt.show()
```

Figure Size + Grid + Background

```python
fig, ax = plt.subplots(figsize=(8,5))

ax.plot(x, y, 'g-', linewidth=2, label='sin(x)')
ax.set_facecolor("#f9f9f9")        # light background
ax.grid(True, linestyle=":", color="gray", alpha=0.7)

ax.set_title("Customized Design", fontsize=18, color="darkred")
ax.set_xlabel("Time", fontsize=12)
ax.set_ylabel("Amplitude", fontsize=12)

ax.legend()
plt.show()
```

Multiple Lines with Styles

```python
y2 = np.cos(x)

plt.plot(x, y, label="sin(x)", color="purple", linewidth=2)
plt.plot(x, y2, label="cos(x)", color="orange", linestyle='--', linewidth=2)

plt.title("Sine & Cosine", fontsize=16, fontweight="bold")
plt.legend(frameon=True, fancybox=True, shadow=True, borderpad=1)
plt.show()
```

Add Annotations

```python
plt.plot(x, y, label="sin(x)")
plt.annotate("Peak", xy=(np.pi/2, 1), xytext=(2,1.3),
arrowprops=dict(facecolor='black', shrink=0.05))
plt.legend()
plt.show()
```
