# 📘 Linear Regression (Quick Notes)

### 🔹 Definition
Linear Regression is a basic ML/statistics algorithm that models the relationship between:
- **Dependent variable (y)** → target/output  
- **Independent variable(s) (x)** → feature(s)  

**Equation:**  
\[
y = \beta_0 + \beta_1 x + \epsilon
\]  
- \( \beta_0 \): Intercept  
- \( \beta_1 \): Slope (coefficient)  
- \( \epsilon \): Error term  

---

### 🔹 Types
- **Simple Linear Regression** → 1 feature  
- **Multiple Linear Regression** → multiple features  

---

### 🔹 Assumptions
1. Linearity → Linear relation between x & y  
2. Homoscedasticity → Constant variance of residuals  
3. Independence → Observations independent  
4. Normality → Residuals ~ Normal distribution  


```python
# Sample dataset
x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # feature
x2 = np.array([10, 20, 70, 40, 50]).reshape(-1, 1)  # feature
y = np.array([2, 4, 5, 4, 5])  # target
```

x → input feature values (independent variable).

.reshape(-1, 1) → changes the shape from (5,) to (5,1) since scikit-learn expects a 2D array for features.

y → target values (dependent variable) that we want to predict

```python
# create model

model = LinearRegression()
```

Creates a Linear Regression model object.

At this point, the model is empty (not trained yet).

```python
model.fit(x, y)
```

Training step → fits the regression line to the given data.

The model learns the slope (β1) and intercept (β0) by minimizing the error (Least Squares Method).

```python
# Predictions
y_pred = model.predict(x2)
y_pred
```

Uses the learned regression equation:


y^	​= β0+β1​x

y_pred contains the predicted values for all x.

```python
# Step 5: Check results
print("Intercept:", model.intercept_)
print("Slope:", model.coef_)
print("Predictions:", y_pred)

# Step 6: Visualization
plt.scatter(x, y, color="blue", label="Data points")
plt.plot(x, y_pred, color="red", label="Regression line")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
# Dataset
data = {
    'Size': [1000, 1500, 2000, 2500, 3000],
    'Bedrooms': [2, 3, 3, 4, 5],
    'Price': [200000, 300000, 400000, 500000, 600000]
}

df = pd.DataFrame(data)
X = df[['Size','Bedrooms']]
y = df['Price']

model = LinearRegression()
model.fit(X, y)

print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)
print("R-squared:", model.score(X, y))

# Prediction
print("Predicted price:", model.predict([[5000,6]])[0])
```
