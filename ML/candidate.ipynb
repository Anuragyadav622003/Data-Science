```python
import pandas as pd  

# --------------------
# Load Data
# --------------------
df = pd.read_csv('CustomCharts-20250927-0328.csv')

# Base Jira URL
base_url = "https://dotcompliancegroup.atlassian.net/browse/"

# Update 'Key' column with full URLs
df['Key'] = base_url + df['Key'].astype(str)

# Save updated CSV
df.to_csv('CustomCharts-20250927-0328.csv', index=False)

# --------------------
# Utility function for filtering by component
# --------------------
def get_keys(df, component_name):
    """Return list of Jira URLs for a given component (case-insensitive)."""
    return df.loc[
        df['Components'].str.strip().str.casefold() == component_name.casefold(),
        'Key'
    ].tolist()

# --------------------
# Get component-specific URLs
# --------------------
dialpad_urls = get_keys(df, 'Dialpad')
vpn_urls = get_keys(df, 'DCG-365')
legacy_urls = get_keys(df, 'Legacy')

print("Dialpad URLs:", dialpad_urls)
print("VPN URLs:", vpn_urls)
print("Legacy URLs:", legacy_urls)

# --------------------
# Component counts
# --------------------
component_counts = df['Components'].value_counts()
print("\nComponent Counts:\n", component_counts)

# Rows where a component occurs more than once
component_counts_more_1 = df[df['Components'].map(component_counts) > 1].reset_index(drop=True)
print("\nRows where Components occur > 1:\n", component_counts_more_1)

# --------------------
# (Optional) Dictionary of all components -> URLs
# --------------------
component_urls = {
    comp: get_keys(df, comp) for comp in df['Components'].dropna().unique()
}
print("\nComponent -> URLs mapping:\n", component_urls)
```
