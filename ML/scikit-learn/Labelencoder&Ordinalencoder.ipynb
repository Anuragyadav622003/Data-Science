Purpose: Converts categorical labels into numeric form.

Use case: When the categorical variable is nominal (no order).

Library: sklearn.preprocessing.LabelEncoder

## 2Ô∏è‚É£ Ordinal Encoding

Purpose: Converts categorical features into numbers with a meaningful order.

Use case: When the categorical variable is ordinal (has order, e.g., low < medium < high).

Library: sklearn.preprocessing.OrdinalEncoder

```python
from sklearn.preprocessing import OrdinalEncoder

# Example data
size = [['small'], ['medium'], ['large'], ['medium'], ['small']]

# Define the order
encoder = OrdinalEncoder(categories=[['small', 'medium', 'large']])

# Fit and transform
size_encoded = encoder.fit_transform(size)

print("Original:", size)
print("Encoded:", size_encoded)
```

## üîπ What is One-Hot Encoding?

Converts categorical values into binary (0/1) columns.

Useful when categories are nominal (no natural order, like fuel = Petrol, Diesel, CNG).

Prevents the model from assuming any numeric relationship between categories.
