### What is the Cross Product?


a=[a1â€‹,a2â€‹,a3â€‹], b=[b1â€‹,b2â€‹,b3â€‹]
aÃ—b = i^(a2â€‹b3â€‹âˆ’a3â€‹b2â€‹)âˆ’j^â€‹(a1â€‹b3â€‹âˆ’a3â€‹b1â€‹)+k^(a1â€‹b2â€‹âˆ’a2â€‹b1â€‹)

âˆ£aÃ—bâˆ£  = âˆ£aâˆ£âˆ£bâˆ£sin(Î¸)

cos(Î¸)  = (a.b/|a||b|)

Î¸  = arccos(a.b/|a||b|)



```python
import numpy as np

# Example vectors
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# -------------------
# 1. Dot product
dot = np.dot(a, b)
print("Dot product:", dot)

# 2. Cross product
cross = np.cross(a, b)
print("Cross product:", cross)

# 3. Norms (magnitudes)
norm_a = np.linalg.norm(a)
norm_b = np.linalg.norm(b)

# 4. Angle using dot product (cosÎ¸ = aÂ·b / |a||b|)
cos_theta = dot / (norm_a * norm_b)
cos_theta = np.clip(cos_theta, -1.0, 1.0)  # avoid floating-point issues
theta_rad = np.arccos(cos_theta)
theta_deg = np.degrees(theta_rad)

print("Angle using dot product (degrees):", theta_deg)

# 5. Angle using cross product (sinÎ¸ = |aÃ—b| / (|a||b|))
sin_theta = np.linalg.norm(cross) / (norm_a * norm_b)
sin_theta = np.clip(sin_theta, -1.0, 1.0)
theta_rad_cross = np.arcsin(sin_theta)
theta_deg_cross = np.degrees(theta_rad_cross)

print("Angle using cross product (degrees):", theta_deg_cross)
```
