```python

from PIL import Image

# Load image (make sure you saved it as numpy_logo.png in your working directory)
img = Image.open("numpy.png").convert("RGBA")  # Convert to RGBA for consistency

# Convert to NumPy array
arr = np.array(img)

# Save as .npy
np.save("numpy_logo.npy", arr)

print("numpy_logo.npy created successfully!")
print("Array shape:", arr.shape)
print("Array dtype:", arr.dtype)
```

```python
import numpy as np
import matplotlib.pyplot as plt

try:
    logo = np.load('numpy_logo.npy')

    # Original logo
    plt.figure(figsize=(12,5))
    plt.subplot(121)
    plt.imshow(logo)
    plt.title("Numpy Logo")
    plt.axis("off")

    # Change color (example: boost blue channel, reduce red/green)
    color_logo = logo.copy().astype(np.float32)
    color_logo[..., 0] *= 0.5   # reduce Red
    color_logo[..., 1] *= 0.5   # reduce Green
    color_logo[..., 2] *= 1.5   # boost Blue
    color_logo = np.clip(color_logo, 0, 255).astype(np.uint8)

    plt.subplot(122)
    plt.imshow(color_logo)
    plt.title("Blue Tint Logo")
    plt.axis("off")

    plt.show()

except FileNotFoundError:
    print("numpy_logo.npy not found")
```
